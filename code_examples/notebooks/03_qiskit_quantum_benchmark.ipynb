# Notebook 03: Qiskit benchmark (visualization only)

All logic lives in `code_examples/src/`: `quantum_benchmark.analyze_problem_from_notebook`, `bundle_labels`, and `benchmark_viz`.

Plots show **inline** below (`%matplotlib inline`) and are also saved under **`code_examples/data/`** (including *Top recommended bundles* with readable package/coverage names).

Run Jupyter with cwd in `code_examples/` or `code_examples/notebooks/`.

````python
%matplotlib inline

import sys
from pathlib import Path

# Allow "from src..." whether kernel cwd is code_examples/ or notebooks/
_here = Path.cwd().resolve()
_code_examples = _here if (_here / "src" / "insurance_model.py").is_file() else _here.parent
if str(_code_examples) not in sys.path:
    sys.path.insert(0, str(_code_examples))

import pandas as pd

from src.insurance_model import load_ltm_instance, subsample_problem
from src.quantum_benchmark import (
    analyze_problem_from_notebook,
    notebook_analysis_to_jsonable,
    print_benchmark_summary,
    resolve_code_examples_root,
    resolve_ltm_data_dir,
)

code_examples_root = resolve_code_examples_root()
ltm_dir = resolve_ltm_data_dir(code_examples_root)
print("code_examples:", code_examples_root)
print("LTM data:", ltm_dir)

ltm_full = load_ltm_instance(ltm_dir)
problem = subsample_problem(ltm_full, n_coverages=5, n_packages=2)
SEED = 7

result = analyze_problem_from_notebook(
    problem,
    save_outputs=True,
    show_plots=True,
    code_examples_root=code_examples_root,
    seed=SEED,
    ltm_dir=str(ltm_dir),
    qaoa_p=2,
    qaoa_maxiter=150,
    qaoa_shots=8192,
    dqi_shots=8192,
)

print_benchmark_summary(result)
display(pd.Series(notebook_analysis_to_jsonable(result)))
print("API-shaped dict keys:", list(notebook_analysis_to_jsonable(result).keys()))
````
